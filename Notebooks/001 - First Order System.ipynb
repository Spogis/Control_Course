```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process

```python
Kp = 20
Tau = 200
sys  = Kp/(Tau*s + 1)
sys
```

```python
# Simulate step response
t, y = control.step_response(sys, T=1500, T_num=1501)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```
