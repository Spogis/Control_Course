## Transfer Functions by Polynomials

```python
num = [30, -3.6, 0.144]
den = [120, 15.4, 0.696, 0.0048]
sys  = control.tf(num, den)
sys
```

```python
# Simulate step response
t, y = control.step_response(sys, T=1500, T_num=1501)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```
