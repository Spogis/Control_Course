```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process

```python
Gs  = 1/(50*s+1)
```

```python
Hs  = 10/(350*s+1)
```

```python
sys = control.feedback(Gs,Hs, sign=-1)
sys 
```

```python
# Simulate step response
t, y = control.step_response(sys)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```
