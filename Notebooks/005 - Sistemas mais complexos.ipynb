```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process

```python
a = 1
b = 2
```

### Bloco em Paralelo

```python
G1 = b/s
```

```python
G2 = a
```

```python
G_Parallel = control.parallel(G1,G2)
G_Parallel
```

## Bloco em Serie

```python
G3 = 1/s
```

```python
G_Serie = control.series(G_Parallel, G3)
G_Serie
```

## Feedback Negativo

```python
sys = control.feedback(G_Serie, 1, sign=-1)
sys
```

## Step Response

```python
# Simulate step response
t, y = control.step_response(sys)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```
