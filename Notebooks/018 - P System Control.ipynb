```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process

```python
Kp = 5
Tau = 300
sys  = Kp/(Tau*s + 1)
```

```python
# Simulate step response
t, y = control.step_response(sys)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```

## P Controller

```python
# Define transfer function of the P controller
Kc = 1.5
Controller = (Kc)
```

## Open Loop

```python
Sys_Control = control.series(Controller, sys)
```

## Closed Loop

```python
# Create closed-loop system
cl_sys = control.feedback(Sys_Control, 1, sign=-1)
```

```python
# Simulate step response
t, y = control.step_response(cl_sys)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('Step response of P controlled system')
plt.grid()
plt.show()
```
