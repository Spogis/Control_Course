```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process  - Underdamped Second Order System

```python
#Natural Period of Oscillation
Tau = 200

#Damping Factor (zeta)
Zeta = 0.5

#Process Gain
K = 1.0
```

```python
A = Tau**2
B = 2*Zeta*Tau
C = 1
sys = K/(A*s**2 + B*s + 1)
sys
```

```python
# Simulate step response
t, y = control.step_response(sys, T=2500, T_num=2501)
```

```python
# Plot step response
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.show()
```
